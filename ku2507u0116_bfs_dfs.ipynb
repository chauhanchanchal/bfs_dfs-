```python
graph = {
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':[],
    'E':['F'],
    'F':[]
}

def dfs(graph, start, goal):
    visited = []
    stack = []

    stack.append(start)
    visited.append(start)
    print(graph)
    print("DFS Traversal:")

    while stack:
        node = stack.pop()
        print(node, end=' ')

        if node == goal:
            print("Goal node found")
            return

        for neighbor in reversed(graph[node]):
            if neighbor not in visited:
                visited.append(neighbor)
                stack.append(neighbor)

    print("Goal node not found")
dfs(graph, 'A', 'D')
```
